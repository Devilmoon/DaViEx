```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./example_data/all_2-Copy1.csv')
```

```python
df2 = pd.DataFrame()
countries = ['Germany', 'Sweden', 'Italy', 'Finland', 'Austria', 'Spain', 'France', 'Denmark']
for index, row in df.iterrows():
    if index % 10000 == 0: print(index)
    if row['Country Name'] in countries:
        df2 = df2.append(row)
```
