```python
import pandas as pd
df1 = pd.read_csv('./data/all.csv')
df2 = pd.read_csv('./data/metadata_country.csv')
df2 = df2.drop(columns=['Country Code', 'IncomeGroup', 'SpecialNotes'])
df2 = df2.rename(index=str, columns={"TableName": "Country Name"})
df_3 = pd.merge(df1, df2, on='Country Name', how='left')
df_3.to_csv('./data/all_2.csv')
df_3
```

```python
import pandas as pd
df_t = pd.read_csv('all_2.csv')
df_t.head()
test = df_t.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 5'])
test.head()

pop = pd.read_csv('population.csv')
pop.head()
```

```python
pop = pop.drop(index=[108])
lname = ''
lregion = ''
for i in range(len(pop)):
    row = pop.iloc[i]
    #print(row["Country Name"], row["Indicator Name"])
    years = row[4:-1].index
    values = row[4:-1]
    name = row["Country Name"]
    if lname != name:
        mask = df_t[df_t["Country Name"] == name]
        region = df_t.at[list(mask.index)[0], 'Region']
        lname = name
        lregion = region
    for i in range(len(values)):
        #print(years[i], values[i])
        new = pd.DataFrame.from_dict([{"Country Name" : row["Country Name"], "Indicator Name" : row["Indicator Name"], "Year" : years[i], "Value" : values[i], "Region": lregion}])
        test = test.append(new, sort=False, ignore_index=True)
test.head()
```

```python
test.tail()
test.to_csv('all_5.csv')
```
