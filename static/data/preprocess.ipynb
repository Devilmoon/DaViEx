```python
import pandas as pd
df1 = pd.read_csv('./data/all.csv')
df2 = pd.read_csv('./data/metadata_country.csv')
df2 = df2.drop(columns=['Country Code', 'IncomeGroup', 'SpecialNotes'])
df2 = df2.rename(index=str, columns={"TableName": "Country Name"})
df_3 = pd.merge(df1, df2, on='Country Name', how='left')
df_3.to_csv('./data/all_2.csv')
df_3
```
