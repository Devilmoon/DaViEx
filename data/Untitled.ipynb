```python
import pandas as pd
df1 = pd.read_csv("GDP_growth.csv")
df2 = pd.read_csv("alternative_and_nuclear_energy.csv")
df3 = pd.read_csv("CO2_emission_kt.csv")
df4 = pd.read_csv("CO2_emissions_per_capita.csv")
df5 = pd.read_csv("electric_power_consumption.csv")
df6 = pd.read_csv("energy_use_kg_of_oil_equivalent_per_capita.csv")
df7 = pd.read_csv("fossil_fuel_energy_consumption.csv")
df8 = pd.read_csv("GDP.csv")
df9 = pd.read_csv("GDP_per_capita.csv")
df10 = pd.read_csv("GDP_per_capita_growth.csv")
df11 = pd.read_csv("GDP_per_unit_of_energy_use.csv")
df12 = pd.read_csv("renewable_elecricity_ouput.csv")
df13 = pd.read_csv("renewable_energy_consumption.csv")
df14 = pd.read_csv("total_greenhouse_gas_emissions.csv")

target = pd.DataFrame(columns=["Country Name","Indicator Name","Year","Value"])
```

```python
target
```

```python
for i in range(len(df1)):
    row = df1.iloc[i]
    #print(row["Country Name"], row["Indicator Name"])
    years = row[4:-1].index
    values = row[4:-1]
    for i in range(len(values)):
        #print(years[i], values[i])
        new = pd.DataFrame.from_dict([{"Country Name" : row["Country Name"], "Indicator Name" : row["Indicator Name"], "Year" : years[i], "Value" : values[i]}])
        target = target.append(new, sort=False, ignore_index=True)
```
